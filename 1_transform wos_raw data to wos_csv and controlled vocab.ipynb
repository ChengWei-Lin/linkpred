```python
data_files = dict()
for i in range(len(data_time_paths)):
    data_files[times[i]] = []
    
    for f_name in os.listdir(data_time_paths[i]):
        data_files[times[i]].append(data_path+times[i]+"\\"+f_name)    
data_files
```

```python
def read_raw_data(files):
    df = pd.DataFrame()
    for file in files:
        df_iter = pd.read_csv(file, sep='\t', index_col=False)
        print("Read {}".format(file))
        print("Number of Rows: {}".format(len(df_iter)))
        df = df.append(df_iter)
        print("Number of All Rows After Appending: {}".format(len(df)))

    return df
```

```python
output_path = path+"\\data_transformation_output\\{}\\wos_csv\\".format(journal)
output_path
```

```python
data = dict()
for key in data_files.keys():
    print('-'*10, key)
    df = read_raw_data(data_files[key])
#     print(output_path+key+".csv")
    df.to_csv(output_path+key+".csv")
    data[key] = df
    print('-'*10, "Saved!", '-'*10, '\n')
```

## stem and lemma
1. data_transformed_stem
2. data_transformed_lemma

```python
import nltk
from nltk.stem.porter import *
stemmer = PorterStemmer()

import en_core_web_sm
nlp = en_core_web_sm.load()

punctuation = ['-', '(', ')', '*', '&', "'s", "s'", "/", "\\", '"', "'", "."]

def generate_lemma_and_stem_text(text, punctuation):
    new_tokens_stem, new_tokens_lemma = '', ''
    if type(text)!=str:
        print('TypeError')
        return '', ''
    else:
        tokens = []
        for word in text.lower().split('; '):            
            for punc in punctuation:
                word = word.replace(punc, '')        

            tokens.append(word)
            
        for token in tokens:
            # token -> 'Journal Impact Factor'
            doc = nlp(token)
            token_stem, token_lemma = '', ''
            # doc -> 'Journal Impact Factor'
            for token in doc:
                # token -> 'Journal'
                token_lema = token.lemma_
                if token.lemma_ == '-PRON-':
                    token_lema = str(token)
                    print('\t\t\tEXCEPTOIN', token_lema) # id, em
                
                
                token_stem += stemmer.stem(token_lema)+' '
                token_lemma += token_lema+' '

            token_stem, token_lemma = token_stem[:-1], token_lemma[:-1] # without last space

            new_tokens_stem += token_stem+'; '
            new_tokens_lemma += token_lemma+'; '
            
        return new_tokens_stem[:-2], new_tokens_lemma[:-2] # without last ;_
```

```python
data_transformed_stem, data_transformed_lemma = dict(), dict()
for i in range(len(times)):
    
    df1 = data[times[i]].copy() # stem
    df2 = data[times[i]].copy() # lemma
    
    list_of_stemmed_text, list_of_lemma_text = [], []
    
    for j, text in enumerate(df1['DE']):
        # print(j, text)
        stemmed_text, lemma_text = generate_lemma_and_stem_text(text, punctuation)
        print(str(j)+'\t' + str(text) + '\n-->\t' + lemma_text + '\n-->\t'+stemmed_text)
        list_of_stemmed_text.append(stemmed_text)
        list_of_lemma_text.append(lemma_text)
        
    df1['DE'] = list_of_stemmed_text
    df2['DE'] = list_of_lemma_text
    
    data_transformed_stem[times[i]] = df1
    data_transformed_lemma[times[i]] = df2
    
    # save to output_path
    df1.to_csv(output_path+"stem\\{}.csv".format(times[i]))
    df2.to_csv(output_path+"lemma\\{}.csv".format(times[i]))
```
