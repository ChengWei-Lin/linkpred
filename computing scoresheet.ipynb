```python
def compute_confusion(evaluation, num_connections):
    
    def add_0_for_cal(X):
        
        C = pd.DataFrame()
        C = C.append([[0, 0, num_connections, -1]])
        C.columns = X.columns
        X = C.append(X)
        
        return X

    X = add_0_for_cal(evaluation)[:-1]
    Y = add_0_for_cal(evaluation)[1:]
    X = X.reset_index()
    X = X.drop(columns=['index'])
    res = Y-X

    return res

df_confusions = dict()

for k in eva.keys():
    evaluation = eva[k]
    num_connections = int(evaluation['tp'][0]+evaluation['fn'][0])
    
    print('-'*10, k, num_connections)
    df_confusion = compute_confusion(evaluation, num_connections)
    df_confusions[k] = df_confusion
    
```

            to be unknown. This still allows for some measures, including
            precision and recall, to be calculated.
